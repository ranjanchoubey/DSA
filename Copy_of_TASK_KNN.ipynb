```python
X = data.drop('Type', axis = 1)
y = data['Type']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test , y_train, y_test = train_test_split(X , y , test_size =0.2 , random_state =40)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import math
k=int(math.sqrt(171))
k
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
score = []
for i in range(1 , k):
  knn = KNeighborsClassifier(n_neighbors = i )
  knn.fit(X_train, y_train)
  y_pred  = knn.predict(X_test)
  score.append(accuracy_score(y_pred , y_test))

```

```python
score
```

```python
sns.lineplot(x = range(1 , k) , y = score)
```

```python
knn = KNeighborsClassifier(n_neighbors = 6 )
knn.fit(X_train, y_train)
y_pred  = knn.predict(X_test)
m = accuracy_score(y_pred , y_test)
m
```

```python
from sklearn.metrics import precision_score , recall_score , f1_score
pre = precision_score(y_pred ,y_test , average ="weighted")
rec = recall_score(y_pred ,y_test , average ="weighted")
f1= f1_score(y_pred , y_test , average ="weighted")
```
